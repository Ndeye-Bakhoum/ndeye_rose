## OFFRE DE TRAVAIL ET SANTE

```python
import numpy as np
from scipy.optimize import fsolve
from scipy.stats import norm
from matplotlib import pyplot as plt
```

Ce projet a pour objectif de déterminer la relation entre l'offre de travail et la santé des ménages en utlisant un modèle microéconomique. La santé est modélisée comme un bien dans lequel les agents investissent via des dépenses de santé, influençant leur bien-être et, indirectement, leurs décisions de travail. En arbitrant entre consommation, loisir, travail et investissement en santé, les ménages déterminent simultanément leur nombre d’heures travaillées et leurs dépenses de santé.


Nous allons d'une part caractériser les décisions optimales d’offre de travail et de demande de santé et d’autre part, le  comparer les comportements des ménages vivant aux États-Unis et en Europe, en mettant en évidence le rôle du prix relatif de la santé et des préférences individuelles. Cette analyse permet ainsi d’illustrer comment les différences institutionnelles et économiques peuvent affecter l’offre de travail via le canal de la santé.



**Les préferences dun agent i sont donnée par**

$$
U_i = \mu \log(c_i) + (1 - \mu) \log(1 - h_i) + \pi(m_i)\phi
$$

Avec                  $$  \pi(m_i) = 1 - \exp(-\alpha_1 - \alpha_1 m_i) $$

où $c_i ≥ 0$ est la consommation, $h_i$ le nombre d’heures travaillées, $m_i$ la dépense de santé et $\pi(m_i)$
la probabilité d’être en bonne santé. Le temps disponible de l’agent est nomalisé à l’unité: $1−hi$
représente donc son loisir. 

**La contrainte budgetaire est**

$$ c_i + pm_i = y_ih_i $$

où $ y_i$ représente le salaire horaire de l’individu et $p$ est le prix relatif de la santé 


### Programme d'optimisation du ménage i

Le ménage aure pour objectif de maximiser son utilité. De ce fait son programme d'optimisation sera de maximiser cette utilité sous la contrainte de son budget.

$$
\max_{c_i,h_i,m_i} \{\mu \log(c_i) + (1 - \mu) \log(1 - h_i) + \phi[1 - e^{-(\alpha_1 + \alpha_0 m_i)}]\}
$$

sous la contrainte budgétaire

$$
c_i = y_i h_i - p m_i
$$

En remplacant $c_i$

$$
\max_{h_i,m_i} \{\mu \log(y_i h_i - p m_i) + (1 - \mu) \log(1 - h_i) + \phi[1 - e^{-(\alpha_1 + \alpha_0 m_i)}]\}
$$


### Condition d'optimalité du ménage i

Les conditions du premier ordre caractérisent les choix optimaux d’offre de travail et de dépenses de santé. 
Elles mettent en évidence l’arbitrage entre consommation, loisir et investissement en santé.

➤ Condition d’optimalité par rapport à $h_i$

$$
\frac{\partial U_i}{\partial h_i}
=
\mu \frac{y_i}{y_i h_i - p m_i}
-
\frac{1-\mu}{1-h_i}
= 0
$$

$$
\mu \frac{y_i}{y_i h_i - p m_i}
=
\frac{1-\mu}{1-h_i}
$$

➤ Condition d’optimalité par rapport à $m_i$

$$
\frac{\partial U_i}{\partial m_i}
=
- \mu \frac{p}{y_i h_i - p m_i}
+ \phi \alpha_0 e^{-(\alpha_1 + \alpha_0 m_i)}
= 0
$$

$$
\mu \frac{p}{y_i h_i - p m_i}
= \phi \alpha_0 e^{-(\alpha_1 + \alpha_0 m_i)}
$$

### Equations à résoudre

les equations permettant de résoudre le programme des ménage sont les suivantes

$$
\begin{cases}
\mu \dfrac{y_i}{y_i h_i - p m_i}
= \dfrac{1-\mu}{1-h_i} \\[0.4cm]
\phi \alpha_0 e^{-(\alpha_1 + \alpha_0 m_i)}
= \mu \dfrac{p}{y_i h_i - p m_i}
\end{cases}
$$



### Procédure d’estimation

#### Étape 1 : Estimation sur les données américaines (US)

On utilise les données américaines pour estimer les paramètres :

$$
\{\phi,\ \alpha_0,\ \alpha_1^{US}\}
$$

à partir de la condition du premier ordre associée à la dépense de santé et des observations :

$$
\{h_i,\ m_i,\ y_i\}
$$

Le prix de la santé est normalisé à l’unité aux États-Unis :

$$
p_{US} = 1
$$

La condition du premier ordre par rapport à \(m_i\) est :

$$
\frac{\mu p_{US}}{y_i h_i - p_{US} m_i}
=
\phi \alpha_0 \exp\left(-\alpha_1^{US} - \alpha_0 m_i\right)
$$

Cette équation permet d’estimer les paramètres $(\phi, \alpha_0 , \alpha_1^{US})$.


#### Étape 2 : Estimation sur les données européennes (EU)

On utilise ensuite les données européennes pour estimer :

$$
\{p,\ \alpha_1^{EU}\}
$$

en supposant que les paramètres de préférence et de technologie de santé sont identiques à ceux estimés sur les données américaines :

$$
\phi^{EU} = \phi^{US}, \qquad \alpha_0^{EU} = \alpha_0^{US}
$$

La condition du premier ordre devient alors :

$$
\frac{\mu p}{y_i h_i - p m_i}
=
\phi \alpha_0 \exp\left(-\alpha_1^{EU} - \alpha_0 m_i\right)
$$

La résolution de cette équation à partir des observations européennes permet d’identifier :

- le prix relatif de la santé en Europe $(p)$,
- le paramètre $(\alpha_1^{EU})$.

---

#### Remarque

L’indice \(i\) correspond à un niveau de revenu.  
L’estimation est répétée pour :

$$
\mu \in \left\{\frac{1}{3},\ \frac{1}{2},\ \frac{2}{3}\right\}
$$

afin d’étudier la sensibilité des résultats aux préférences des ménages.

### Résolution numérique

Dans cette partie, nous allons creer des fonctions permettant de traduire le programme de maximisation du ménage en intégrant la contrainte budgétaire dans la fonction d’utilité, ce qui permet de réduire le problème à un choix non contraint des heures de travail et des dépenses de santé. Les conditions du premier ordre associées à ces deux décisions sont ensuite dérivées et traduites sous la forme d’un système d’équations non linéaires.

Les solutions analytiques n’étant pas disponibles, le système est résolu numériquement afin d’obtenir, pour chaque niveau de revenu, les choix optimaux de travail et de dépenses de santé du ménage.

```python
# Fonctions du modèle

def pi(m, alpha0, alpha1):
    """Probabilité d'être en bonne santé"""
    return 1 - np.exp(-alpha1 - alpha0 * m)

def first_order_system(vars, y, mu, phi, alpha0, alpha1, p):
    """Système d'équations pour h et m"""
    h, m = vars
    c = y * h - p * m
    if c <= 0 or h <= 0 or h >= 1 or m < 0:
        return [1e6, 1e6]  
    eq1 = (mu * y) / c - (1 - mu) / (1 - h)
    eq2 = - (mu * p) / c + phi * alpha0 * np.exp(-alpha1 - alpha0 * m)
    return [eq1, eq2]

def solve_h_m(y, mu, phi, alpha0, alpha1, p, h0=0.5, m0=0.5):
    """Résolution numérique de h et m"""
    solution = fsolve(first_order_system, [h0, m0], args=(y, mu, phi, alpha0, alpha1, p), maxfev=2000)
    return solution
```

### Données

Nous allons utilisé des simulées des données sur le revenu, les dépenses de santé et le temps de travail pour estimer les différents paramétres de notre modèle. Nous nous basons sur le théorie économique comme du fait que les revenus sont en général plus élevés au US et les heures de travail plus élevés UE.

```python
np.random.seed(42)

N = 15
mu_values = [1/3, 1/2, 2/3]

# Revenus US et EU
y_US = np.random.uniform(15, 40, N)
y_EU = y_US * np.random.uniform(0.8, 0.95, N)  

# Heures travaillées (fonction croissante du revenu)
h_US_obs = 0.3 + 0.005 * (y_US - y_US.min())  # 0.3 -> 0.45
h_EU_obs = h_US_obs + 0.02  

# Dépenses santé (fonction croissante du revenu)
m_US_obs = 1.0 + 0.05 * (y_US - y_US.min())
m_EU_obs = m_US_obs * 1.05  
```

### Estimation des paramétres US

Dans cette partie nous estimons les paramétres phi(le bonheur d'etre en santé), alpha0(la productivité du dépense de médicament sur le probabilité d'etre en bonne santé) et alpha1(la norme social en terme de risque dans l'économie) partir des données américaines en utilisant une méthode de minimisation numérique. Notre fonction va mesurer l’écart entre les choix observés des ménages (heures travaillées et dépenses de santé) et les choix prédits par le modèle pour un ensemble donné de paramètres. Pour chaque ménage, le programme du ménage est résolu numériquement afin d’obtenir les valeurs optimales prédites de travail et de dépenses de santé, en supposant que le prix de la santé est normalisé à un aux États-Unis.

```python
# Étape 1 : Estimation des paramètres US

# Nous faisons une estimation simplifiée par minimisation de l'erreur entre observé et prédit

def loss_US(params, y, h_obs, m_obs, mu):
    phi, alpha0, alpha1 = params
    loss = 0
    for i in range(len(y)):
        h_pred, m_pred = solve_h_m(y[i], mu, phi, alpha0, alpha1, p=1.0)
        loss += (h_pred - h_obs[i])**2 + (m_pred - m_obs[i])**2
    return loss

estimated_params_US = {}
for mu in mu_values:
    res = minimize(loss_US, x0=[1.0, 0.5, 0.5], args=(y_US, h_US_obs, m_US_obs, mu),
                   bounds=[(0.1, 5), (0.01, 2), (0.0, 2)], method='L-BFGS-B')
    estimated_params_US[mu] = res.x
    print(f"US params for mu={mu:.3f}: phi={res.x[0]:.3f}, alpha0={res.x[1]:.3f}, alpha1_US={res.x[2]:.3f}")
```

Nous pouvons constater plus le mu augmente plus le bonheur d'etre en bonne santé augmente de meme que la productivité du dépense de médicament sur le probabilité d'etre en bonne santé et la norme social

### Estimation des paramétres UE

Dans cette partie, nous considérons que es paramétres phi(le bonheur d'etre en bonne santé), alpha0(la productivité du dépense de médicament sur le probabilité d'etre en bonne santé) ne change pas entre les pays. De ce fait nous estimons le prix des soins et alpha1 la norme social en terme de risque dans l'économie. En effet ce dernier permet de capter les différences exogènes d’état de santé entre les ménages européens et américains. L’estimation est répétée pour chaque valeur de 𝜇


```python
# Étape 2 : Estimation des paramétres UE

def loss_EU(params, y, h_obs, m_obs, mu, phi, alpha0):
    p, alpha1_EU = params
    loss = 0
    for i in range(len(y)):
        h_pred, m_pred = solve_h_m(y[i], mu, phi, alpha0, alpha1_EU, p)
        loss += (h_pred - h_obs[i])**2 + (m_pred - m_obs[i])**2
    return loss

estimated_p_EU = {}
for mu in mu_values:
    phi_US, alpha0_US, _ = estimated_params_US[mu]
    bounds_EU = [(0.1, 1.0), (0.0, 5.0)]  # p <= 1
    res_EU = minimize(loss_EU, x0=[0.5, 0.5], args=(y_EU, h_EU_obs, m_EU_obs, mu, phi_US, alpha0_US),
                      bounds=bounds_EU, method='L-BFGS-B')
    estimated_p_EU[mu] = res_EU.x
    print(f"EU params for mu={mu:.3f}: p={res_EU.x[0]:.3f}, alpha1_EU={res_EU.x[1]:.3f}")
```

Aprés estimation nous pouvons observer le prix des soins en Europe est inférieur à celui des Etats-Unis ce qui est cohérent et aussi on note que au fur et à mesure que mu augmente le prix des soins diminue. De ce fait il joue un role important dans l'ecart des prix de santé entre les US et UE

### Resultat final

```python
# Cmparaison US et UE

mu = 0.5

# On récupère les paramètres estimés
phi_US, alpha0_US, alpha1_US = estimated_params_US[mu]
p_EU, alpha1_EU = estimated_p_EU[mu]

print(f"{'Revenu':>6} | {'h_US':>6} {'m_US':>6} | {'h_EU':>6} {'m_EU':>6}")
print("-"*36)

# On compare les individus avec le même revenu
for i in range(len(y_US)):
    y = y_US[i] 
    # Prédictions US
    h_US, m_US = solve_h_m(y, mu, phi_US, alpha0_US, alpha1_US, p=1.0)
    # Prédictions EU
    h_EU, m_EU = solve_h_m(y, mu, phi_US, alpha0_US, alpha1_EU, p_EU)
    
    print(f"{y:6.2f} | {h_US:6.3f} {m_US:6.3f} | {h_EU:6.3f} {m_EU:6.3f}")
```

Nous pouvons constater que pour deux individus ayant le meme revenu l'un vivant au US et l'autre UE, le temps de travail effectué par la personne résidant aux Etats Unis est supérieur à celui résidant en Europe.
